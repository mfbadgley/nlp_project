```python
#assign collection to variable, use list to iterate object cursor
temp = list(collection.find())
#assign to dataframe,make sure temp is indexed correctly 
data = pd.DataFrame({'data':temp[1]})#here is 2 because data was inserted 2nd into collection, i.e. is 2nd document
data.index.name = 'url'#make the index URL
data.reset_index(level=0, inplace=True)#reset to have integer index, inplace
data.head()
all_keys = temp[0]['all_keys']#keys inserted first, returns a list of the href/urls, put into variable all_keys
# a function to put href/urls back in..
def revindex(row):
    try:
        return all_keys[int(row['url'])]#returns the key indexed at this dataframe row#
    except:
        return '_id'
data['url'] = data.apply(revindex, axis=1)#url column will apply the 'revindex' function, row by row, and populate with hrefs/urls
```

```python
#pull out the tag that, mostly, contains the date...
from bs4 import BeautifulSoup
def date_tag(row):
    text=row['data']
    soup=BeautifulSoup(text,'lxml')
    vs =soup.findAll('p')[0].text
    return pd.Series(vs)
data_nflx_three[['datetag']] = data_nflx_three.apply(date_tag, axis=1)

#in two step process, pull out ticker...first find the a tag, that mostly contains the right ticker, but since it's
#not always, we then find the mode (most occurring) and then with a 2nd function, apply that mode into new column, and 
#drop the first column..prob a more efficient way to do this, but..
def ticker(row):
    text=row['data']
    soup=BeautifulSoup(text,'lxml')
    tick=soup.findAll('a')[0].text
    return pd.Series(tick)
data_nflx_three[['tick']] = data_nflx_three.apply(ticker, axis=1)
def tickers(row):
    tick=data_nflx_three.tick.mode()
    return pd.Series(tick)
data_nflx_three[['ticker']]=data_nflx_three.apply(tickers, axis=1)
data_nflx_three.drop(['tick'], axis=1, inplace=True)
data_nflx_three.head(20)
```

```python
#split management presentation and question and answer session, some have multiple occurrences of the split phrase, so
#must deal with that, in this case designating first occurence; 'nparts' tells you have many segments
#it splits into..
import re
def splitqa(row):
    datas = row['data'].lower()
    try:
        #first, last = datas.split('question-answer-session')
        #values = re.split("question-answer-session|question and answer session|Question-And-Answer Session", datas)
        #first = values[0]
        #last = ' '.join(values[1:])
        first, last = re.split("question-answer-session|question and answer session|question-and-answer session|operator instructions", datas, 1)#1 split at max to occure
    except:
        first = datas
        last = ''
    nparts = len((re.split("question-answer-session|question and answer session|question-and-answer session|operator instructions", datas)))
    return pd.Series([first, last, nparts])
data_nflx_three[['beforeqa', 'afterqa', 'nparts']] = data_nflx_three.apply(splitqa, axis=1)
data_nflx_three
```

```python
#lets pull the date using regex 
import re
def find_date(row):
    datetag=row['datetag']
    try:
        regex = r"([a-zA-Z]+) (\d+),\s\d{4}"
        x=re.search(regex, datetag)
        z=x.group()
    except:
        z=' '
    return pd.Series(z)
data_nflx_three[['date']] = data_nflx_three.apply(find_date, axis=1)
data_nflx_three
```

```python
#remove html and rename 'afterqa'/'beforeqa' columns, remove html after split as html tags help with split..
from bs4 import BeautifulSoup
def removehtml(row):
    text=row['afterqa']
    soup=BeautifulSoup(text,'html.parser')
    vs =soup.text.replace('\n', '')
    return pd.Series(vs)
def removehtml_2(row):
    text=row['beforeqa']
    soup=BeautifulSoup(text,'html.parser')
    vs =soup.text.replace('\n', '')
    return pd.Series(vs)
data_nflx_three[['QA']] = data_nflx_three.apply(removehtml, axis=1)
data_nflx_three[['MP']] = data_nflx_three.apply(removehtml_2, axis=1)
data_nflx_four=data_nflx_three.drop(['afterqa','beforeqa'], axis=1)
data_nflx_four.head(50)
```

```python

#strptime is short for "parse time" (parse (convert) string to datetime object)..where strftime is for 
#"formatting time" (create formatted string for given time/date/datetime object according to specified format). 
#That is, strptime is the opposite of strftime though they use, conveniently, the same formatting specification
import datetime
import holidays
from datetime import timedelta, date
from bdateutil import isbday, relativedelta
date_df = sorted_df.filter(['date'], axis=1)#pull in date field from sorted_df into new dataframe

#convert string date format to one that can be used later in stock price scrape (YYYY-MM-DD), also check if initial
#date is valid biz day/non holiday
def convert_date_string(row):
    import datetime
    start = row['date']
    start_date = datetime.datetime.strptime(str(start), '%B %d, %Y').date()#parse the date's original format
    if isbday(start_date, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
        start_date
    else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
        start_date=(datetime.datetime.strptime(str(start_date),'%Y-%m-%d')+ relativedelta(bdays=+1, holidays=holidays.US()))
    return pd.Series(start_date.strftime('%Y-%m-%d'))#format it this way
date_df['date_string'] = date_df.apply(convert_date_string, axis=1)

#add a date column of conf call date plus one day
def convert_date_string_plus1(row):
    try:
        date = row['date_string']
        date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')+timedelta(days=1))#add one day to the date_string
        if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
            date_plus
        else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
            date_plus=(datetime.datetime.strptime(str(date_plus),'%Y-%m-%d')+ relativedelta(bdays=+1, holidays=holidays.US()))
            date_plus.strftime('%Y-%m-%d')#format it this way
        return pd.Series(date_plus)
    except:#here if it's not workng, we convert to pydatetime obj (which wil give us some vales to extract, 
        #pull out the year, date, month, hour and retimestamp, then try the process again (of adding days)..sems to work...otherwise was continually getting caught up on 'unconverted data remains'
        dated = pd.to_datetime(row['date'])#first convert string to datetime obj
        #date=dated.to_pydatetime()#then to this pytdatetime (str won't convert right to this) to pull out pieces
        #date = str(row['date_string']).to_pydatetime()
        #a=date.year
        #b=date.month
        #c=date.day
        #d=date.hour
        #date_df.at[row,'date']= None
        #date_df.at[row, 'date'] = pd.Timestamp(a, b, c, d)
        date_plus = (datetime.datetime.strptime(str(dated), '%Y-%m-%d %H:%M:%S')+timedelta(days=1))#add one day to the date_string
        if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
            date_plus
        else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
            date_plus=(datetime.datetime.strptime(str(date_plus),'%Y-%m-%d %H:%M:%S')+ relativedelta(bdays=+1, holidays=holidays.US()))
            date_plus.strftime('%Y-%m-%d')
#         bad_counter+=1
#         bad_values.append(row.name)
        return pd.Series(date_plus)
date_df['date_string_plus1'] = date_df.apply(convert_date_string_plus1, axis=1)

def convert_date_string_plus30(row):
    #try:
    date = row['date_string']
    date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')+timedelta(days=30))#add 30 days
    if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
         return date_plus
    else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
        date_plus=date_plus+relativedelta(bdays=+1, holidays=holidays.US())
        return date_plus
    return pd.Series(date_plus)
date_df['date_string_plus30'] = date_df.apply(convert_date_string_plus30, axis=1)

#add a date column of conf call date plus 60 days
def convert_date_string_plus60(row):
    date = row['date_string']
    date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')+timedelta(days=60))#add 60 days
    if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
         return date_plus
    else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
        date_plus=date_plus+relativedelta(bdays=+1, holidays=holidays.US())
        return date_plus
    return pd.Series(date_plus)
date_df['date_string_plus60'] = date_df.apply(convert_date_string_plus60, axis=1)

def convert_date_string_plus90(row):
    date = row['date_string']
    date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')+timedelta(days=90))#add 60 days
    if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
         return date_plus
    else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
        date_plus=date_plus+relativedelta(bdays=+1, holidays=holidays.US())
        return date_plus
    return pd.Series(date_plus)
date_df['date_string_plus90'] = date_df.apply(convert_date_string_plus90, axis=1)

def convert_date_string_plus180(row):
    date = row['date_string']
    date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')+timedelta(days=180))#add 60 days
    if isbday(date_plus, holidays=holidays.US())==True:#if it is a biz day and not holiday, use it
         return date_plus
    else:#this will take date input and add to the next biz day (non-weekend), taking into account US holidays...
        date_plus=date_plus+relativedelta(bdays=+1, holidays=holidays.US())
        return date_plus
    return pd.Series(date_plus)
date_df['date_string_plus180'] = date_df.apply(convert_date_string_plus180, axis=1)

        
#add the ticker column
date_df['ticker'] = sorted_df.filter(['ticker'], axis=1)
date_df.drop(['date'], inplace=True, axis=1)
date_df
```

```python
#date dataframe for event study around earnings...

#benchmark
tick='SPY'

earn_surp_date = date_df.filter(['date_string', 'date_string_plus1', 'ticker'], axis=1)
good_friday=['2001-04-13', '2002-03-29', '2003-04-18', '2004-04-09', '2005-03-25', '2006-04-14', '2007-04-06', '2008-03-21', '2009-04-10', '2010-04-02'
            ,'2011-04-22', '2012-04-06', '2013-03-28', '2014-04-18', '2015-04-03', '2016-03-25', '2017-04-14']

#find the biz day prior to earnings conf call (ie. earnings release) day
def convert_date_string_minus1(row):
    date = row['date_string']
    date_plus = (datetime.datetime.strptime(date, '%Y-%m-%d')-timedelta(days=1))#subtract 1 day
    if isbday(date_plus, holidays=holidays.US()) and date_plus.strftime('%Y-%m-%d')not in good_friday:#if it is a biz day and not holiday, use it
         return date_plus
    elif (date_plus+relativedelta(bdays=-1, holidays=holidays.US())).strftime('%Y-%m-%d') in good_friday: #so assuming we have rolled back in to a non-biz day/holiday, if we use relativedelta -1, do we end up in the good friday list? if so roll back one more day
        date_plus=date_plus+relativedelta(bdays=-2, holidays=holidays.US())
        return date_plus     
    else:#if we havent' rolled back into good friday, above,  then use the standard 1 day roll-back;  
        date_plus=date_plus+relativedelta(bdays=-1, holidays=holidays.US())
        return date_plus
    return pd.Series(date_plus)
earn_surp_date['date_string_minus1'] = date_df.apply(convert_date_string_minus1, axis=1)

#find the stock price for day prior to earnings
def stock_price(row):
    from yahoo_fin.stock_info import get_data
    date=row['date_string_minus1']
    ticker=row['ticker']
    try:
        stock_price=get_data(ticker, start_date=date, end_date=date)
        stock_close=(stock_price['close'][0]).round(2)
        return pd.Series(stock_close)
    except:
        return ''
earn_surp_date['stock_price_minus1'] = earn_surp_date.apply(stock_price, axis=1)

#find the stock price for day after earnings
def stock_price(row):
    from yahoo_fin.stock_info import get_data
    date=row['date_string_plus1']
    ticker=row['ticker']
    stock_price=get_data(ticker, start_date=date, end_date=date)
    stock_close=(stock_price['close'][0]).round(2)
    return pd.Series(stock_close)
earn_surp_date['stock_price_plus1'] = earn_surp_date.apply(stock_price, axis=1)

#get index price for day prior, to calc excess return
def index_price1(row):
    from yahoo_fin.stock_info import get_data
    date=row['date_string_minus1']
    ticker=tick
    stock_price=get_data(ticker, start_date=date, end_date=date)
    stock_close=(stock_price['close'][0]).round(2)
    return pd.Series(stock_close)
earn_surp_date['index_price_minus1'] = earn_surp_date.apply(index_price1, axis=1)

#get index price for day after
def index_price1(row):
    from yahoo_fin.stock_info import get_data
    date=row['date_string_plus1']
    ticker=tick
    stock_price=get_data(ticker, start_date=date, end_date=date)
    stock_close=(stock_price['close'][0]).round(2)
    return pd.Series(stock_close)
earn_surp_date['index_price_plus1'] = earn_surp_date.apply(index_price1, axis=1)

#calc excess return for the period around earnings t-1 to t+1; 
earn_surp_date['stock_period_return']=((earn_surp_date['stock_price_plus1']-earn_surp_date['stock_price_minus1'])/earn_surp_date['stock_price_minus1']).round(4)
earn_surp_date['index_period_return']=((earn_surp_date['index_price_plus1']-earn_surp_date['index_price_minus1'])/earn_surp_date['index_price_minus1']).round(4)
earn_surp_date['excess_return']=earn_surp_date['stock_period_return']-earn_surp_date['index_period_return']
earn_surp_date
```
